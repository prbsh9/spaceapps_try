# From nasa climate data

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python

url = "https://data.giss.nasa.gov/gistemp/graphs/graph_data/Global_Mean_Estimates_based_on_Land_and_Ocean_Data/graph.txt"
data = requests.get(url).text
formatted_data = data.split("-----------------\n")
formatted_data_list = formatted_data[2].split("\n")
```

```python
data = []
for each_data in formatted_data_list:
  d = each_data.split("     ")
  data.append(d)

# final_data

data = data[0:-1]
# data

for i in range(len(data)):
    for j in range(len(data[i])):
        data[i][j] = float(data[i][j])

# Create a Pandas DataFrame
df = pd.DataFrame(data, columns=["Year No", "Smoothing", "Lowess(5)"])
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt


# Extract the features (Year) and target variable (Lowess(5))
X = df[['Year No']]
y = df['Smoothing']

# Create a Linear Regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Make predictions
predictions = model.predict(X)

# Plot the data points and the regression line
plt.scatter(X, y, label='Data')
plt.plot(X, predictions, color='red', linewidth=2, label='Linear Regression')
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (C)')
plt.title('Linear Regression: Year vs Temperature Anomaly (C)')
plt.legend()
plt.show()

# Print the coefficients of the linear regression equation
print('Intercept:', model.intercept_)
print('Coefficient:', model.coef_[0])
```

```python
new_data_point = [[2123.0]]  # Replace this with the year for which you want to make a prediction

# Use the model to make predictions
predicted_value = model.predict(new_data_point)

# Print the predicted value
print('Predicted Lowess(5) for 2025:', predicted_value[0])
```
